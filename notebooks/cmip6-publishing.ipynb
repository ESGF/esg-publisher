How To for CMIP6
=========
Preparation.  have run esgmapfiles (in esgprep) to make mapfiles
--------------
Step 1. esgpublish to postgres database
--------------


```python
 # map files as generated by esgmapfile are in a ./mapfiles subdirectory of the current working directory
 # [--set-replica] is used only at replica centers
esgpublish --project cmip6 --map ./mapfiles [--set-replica]
```

Step 2. esgpublish to thredds
----------------------------------

```python
 # [--set-replica] is used only at replica centers
esgpublish --project cmip6 --map ./mapfiles --service fileservice --noscan --thredds [--set-replica]
```

Step 3. esgpublish to index node
----------------------------------

```python
 # [--set-replica] is used only at replica centers
esgpublish --project cmip6 --map ./mapfiles --noscan --publish [--set-replica]
```
